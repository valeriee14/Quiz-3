```
import seaborn as sns
sns.set()

n = 10000
xplot = np.linspace(53,81,n)
eta = beta0 + beta1 * xplot
y = 1 / (1 + np.exp(-eta))

sns.scatterplot('Temperature', 'Failure', data = df, hue = 'Failure')
sns.lineplot(xplot,y)
```
